```python
#Define a Função para listar os arquivos e sub-diretórios.
def list_files_path(path):
    path_files = []
    for p, _, files in os.walk(os.path.abspath(path)):
        for file in files:
            path_files.append(os.path.join(p, file))
    
    return path_files
```

```python
# Define a Função para Cria um diretório, a partir de uma lista, se não existir
def criar_diretorio_list(list_dir, path):
    for dir in list_dir:
        novo_dir = os.path.join(os.path.abspath(path), str(dir))
        
        if not os.path.exists(novo_dir):
            #print(novo_dir)
            os.mkdir(novo_dir)
```

```python
# Define a Função para Cria um diretório, se não existir
def criar_diretorio(path, recursivo=False):
    if not os.path.exists(path):
        #print(novo_dir)
        if recursivo:
            os.makedirs(path)
        else:
            os.mkdir(path)
```

```python
# Define a Função que retorna o ano do arquivo.
def ano_arquivo(file):
    return datetime.datetime.fromtimestamp(os.stat(file).st_mtime).year
```

```python
# Denfine a função que retorna um dicionário com a quantidade arquivos por diretório e a frequncia dos anos. 
def qtd_arquivo_dir(path):
    
    qtd_arq_dir = {}

    for root, dirs, files in os.walk(path):

        freq_ano = {}

        #print(root)

        split_dir = root.split('\\')
        nome_dir = split_dir[len(split_dir)-1]

        anos_arq = [ano_arquivo(os.path.join(root, f)) for f in files]

        for ano in anos_arq:
            freq_ano[ano] = anos_arq.count(ano)


        qtd_arq_dir[nome_dir] = len(files), freq_ano, root

        #print('----------------------------')

    return qtd_arq_dir
```
